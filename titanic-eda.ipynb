```python
df_class1=df[df['Pclass']==1]
df_class1
```

```python
count1 = df_class1[(df_class1['Embarked'] == 'C') & (df_class1['Survived'] == 0)].shape[0]
```

```python
print("Number of first-class passengers who embarked from 'C' and did not survive:", count1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Embarked', data=df_class1, palette='viridis')
plt.title('Count of First-Class Passengers by Embarkation Point')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.show()
```

```python
df_class1 = df[df['Pclass'] == 1]


embarked_survived_counts = df_class1.groupby(['Embarked', 'Survived']).size().unstack().fillna(0)


embarked_survived_counts.plot(kind='bar', stacked=True, figsize=(10, 6), color=['red', 'green'])
plt.title('Survival Counts by Embarkation Point for First-Class Passengers')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['Not Survived', 'Survived'])
plt.xticks(rotation=0)
plt.show()
```
